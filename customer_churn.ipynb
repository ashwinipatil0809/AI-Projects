```python
#scaling on above other column which are not in 1 or 0 bcz of we have to same scale
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
X[['tenure','MonthlyCharges','TotalCharges']]=sc.fit_transform(X[['tenure','MonthlyCharges','TotalCharges']])
```

```python
# cv train test split
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2,random_state=11)
```

```python
# step -5 model training
#ANN
```

```python
#Architecture of ANN
# Architecture of ANN
# input layer >> input shape >> no of features >> no of pixel
# hidden layer >> no of neurons ,activation function 
# hidden layer is aslo know as Dense layer
# ouput layer  >> no of neurons ,activation function


# output layer >> no of neurons >> should be no of categories in target variable
# in case of two categories ,can write 2 or 1
```

```python
model=keras.Sequential([
    keras.layers.Input(shape=(19,)), #input layer
    keras.layers.Dense(16,activation="sigmoid"), #hidden layer 
    keras.layers.Dense(8,activation="sigmoid"), # hidden layer 
    keras.layers.Dense(1,activation="sigmoid") # output layer
])
```

```python
model.compile(optimizer="adam",loss="binary_crossentropy",metrics=["accuracy"])  #model crate
```

```python
model.summary()
```

```python
model.fit(X_train,y_train,epochs=200)
```

```python
#training data accuracy >> 82%
```

```python
#steps-6 model testing
```

```python
model.evaluate(X_test,y_test)
```

```python
y_pred=model.predict(X_test)
```

```python
y_pred1=[]
for i in y_pred:
    if i[0]>0.5:
        y_pred1.append(1)
    else:
        y_pred1.append(0)
```
